```python
def parse_bible_verses(file_path):
    parsed_versus = []
    with open(file_path, 'r') as file:
        for line in file:
            # print(f"parsing line: {line}")
            # verse_text, reference = line.split('--')
            verse_text, reference = line.rsplit('--', 1)
            verse_text = verse_text.strip()

            book, chapter_and_verse = reference.strip().rsplit(' ', 1)
            chapter, verse_number = chapter_and_verse.split(':')

            parsed_versus.append({
                'verse_text': verse_text,
                'book': book,
                'chapter': chapter,
                'verse_number': verse_number
            })
    return parsed_versus
```

```python
import csv
def write_parsed_verses_to_csv(parsed_verses, output_file):
    with open(output_file, 'w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=['verse_text', 'book', 'chapter', 'verse_number'])
        writer.writeheader()
        writer.writerows(parsed_verses)
```
