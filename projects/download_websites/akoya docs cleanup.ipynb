```python
import os
from bs4 import BeautifulSoup
save_directory = "cleaned-up"
def modify_html_files(directory_path):
    # Iterate over all files in the directory
    for filename in os.listdir(directory_path):
        # Ensure we're only processing .html files
        if filename.endswith(".html"):
            print("cleaning: ", filename)
            filepath = os.path.join(directory_path, filename)

            # Read the content of the file
            with open(filepath, 'r', encoding="utf-8") as file:
                content = file.read()

            # remove redundant tags
            soup = BeautifulSoup(content, 'html.parser')
            for script in soup(["header"]):
                script.extract()  # Remove the tag

            for div in soup.find_all("div", class_=lambda x: x and "Footer" in x):
                div.decompose()

            for div in soup.find_all("div", class_=lambda x: x and "PageThumbs" in x):
                div.decompose()

            for div in soup.find_all("div", class_=lambda x: x and "UpdatedAt" in x):
                div.decompose()

            for div in soup.find_all("div", class_=lambda x: x and "hub-loading-indicator" in x):
                div.decompose()
            #
            # for div in soup.find_all("div", class_=lambda x: x and "ng-non-bindable" in x):
            #     div.decompose()
            #
            for div in soup.find_all("div", class_=lambda x: x and "hub-footer" in x):
                div.decompose()
            #
            for div in soup.find_all("div", class_=lambda x: x and "ssr-top" in x):
                div.decompose()
            #
            for div in soup.find_all("div", id="hub-container"):
                div.decompose()
            #
            for div in soup.find_all("div", id="ssr-end"):
                div.decompose()

            # Remove the class attribute from all elements
            for element in soup.find_all(True):  # This will select all elements
                if 'class' in element.attrs:
                    del element['class']
            for element in soup.find_all(True):  # This will select all elements
                if 'ng-class' in element.attrs:
                    del element['ng-class']
            for element in soup.find_all(True):  # This will select all elements
                if 'style' in element.attrs:
                    del element['style']

            # Create a new <h1> tag with the filename
            h1_tag = soup.new_tag("h1")
            section_name = filename.replace("docs.akoya.com_docs_", "").replace(".html", "").replace("-", " ").replace("_", " ")
            h1_tag.string = section_name
            soup.body.insert(0, h1_tag)

            # Write the modified content back to the file
            new_path = save_directory + "/" + filename
            with open(new_path, 'w', encoding="utf-8") as file2:
                file2.write(str(soup))

# Call the function
directory_path = "akoya-docs-retrieved"
modify_html_files(directory_path)
```
