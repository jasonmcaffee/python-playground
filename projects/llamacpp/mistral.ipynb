````python
from llama_cpp import Llama
import time 
max_tokens = 32000
# llm = Llama(model_path="/Volumes/4Tera_SSD_2022/llm_models/mistral-7b-instruct-v0.2.Q5_K_M.gguf", n_ctx=max_tokens)
llm = Llama(model_path="/Users/jmcaffee/llm_models/mistral-7b-instruct-v0.2.Q5_K_M.gguf", n_ctx=max_tokens)
llm.verbose = False # don't print a bunch of stuff
````

````python
def stream_prompt(question):
    start_time_seconds = time.time()
    stream = llm.create_chat_completion(
        max_tokens=max_tokens,
        messages = [
            {
                "role": "system",
                "content": """
                You are an assistant who perfectly answers questions related to documentation provided.
                If the documentation does not contain the information needed to answer the question, you simply say that you do not know, and do not use any other knowledge to try and answer the question.
                The documentation is provided inside of a <documentation> xml tag.  There can be many <document> entries in the documentation, some of which may not be applicable to the question asked.
                """
            },
            {
                "role": "user",
                "content": question
            }
        ],
        stream = True,
    )


    for output in stream:
        choices = output['choices']
        choice = choices[0]
        delta = choice['delta']
        if 'content' in delta:
            print(delta['content'], end='')
    print()

    elapsed_time_seconds = time.time() - start_time_seconds
    print(f'answer generated in {elapsed_time_seconds} seconds')
````
